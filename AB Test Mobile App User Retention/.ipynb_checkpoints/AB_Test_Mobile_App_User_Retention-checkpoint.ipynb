# Background

**Client:** <p><a href="http://tactile.dk">Tactile Entertainment</a> wants to increase player retention for its hugely popular mobile puzzle game, <a href="https://www.facebook.com/cookiecatsgame">Cookie Cats</a>. Cookie Cats is a classic "connect three"-style puzzle game where the player must connect tiles of the same color to clear the board and win the level. It also features singing cats.</p>
<p>A short demo:</p>
<p><a href="https://youtu.be/GaP5f0jVTWE"><img src="https://s3.amazonaws.com/assets.datacamp.com/production/project_184/img/cookie_cats_video.jpeg" style="width: 500px"></a></p>
<p>As players progress through the levels of the game, they will occasionally encounter gates that force them to wait a non-trivial amount of time or make an in-app purchase to progress. In addition to driving in-app purchases, these gates serve the important purpose of giving players an enforced break from playing the game, hopefully resulting in that the player's enjoyment of the game being increased and prolonged.</p>
<p><img src="https://s3.amazonaws.com/assets.datacamp.com/production/project_184/img/cc_gates.png" alt=""></p>

**Task:** <p>Tactile Entertainment wants to decide where to place the first gate. Initially the first gate was placed at level 30. In this notebook, I will analyze an AB-test that moved the first gate in Cookie Cats from level 30 to level 40. Specifically, I will analyze whether placing the first gate at level 30 or level 40 will lead to higher player retention.</p>

# Overview of the AB-test data

**Ideal Data (not necessarily available):**
- Which version the player was assigned to (gate 30 or gate 40)
- Confirmation that the player should be included in the AB test
    - i.e. whether the player reached the gate
- A player retention metric

**Available Data:**
<p>Tactile Entertainment has provided data from the 90,189 players who installed the game while the AB-test was running. The variables are:</p>
<ul>
<li><code>userid</code> - a unique number that identifies each player.</li>
<li><code>version</code> - whether the player was put in the control group (<code>gate_30</code> - a gate at level 30) or the group with the moved gate (<code>gate_40</code> - a gate at level 40).</li>
<li><code>sum_gamerounds</code> - the number of game rounds played by the player during the first 14 days after install.</li>
<li><code>retention_1</code> - did the player come back and play <strong>1 day</strong> after installing?</li>
<li><code>retention_7</code> - did the player come back and play <strong>7 days</strong> after installing?</li>
</ul>

**Challenge Posed by Lack of Available Data:**
<p>There is no way to confirm whether a user reached their assigned gate and thus should be included in the AB test analysis. However, the <code>sum_gamerounds</code> variable will be helpful for identifying some, but not all, of the players who should not be included in the AB test analysis: A player who has played fewer than 29 rounds could not have reached either gate.</p>

```python
df["irrelevant"] = df["sum_gamerounds"].apply(lambda x: 1 if x < 29 else 0)
```

```python
# Number of players in each of the two AB groups (control and treatment).
df[df["irrelevant"]==0].groupby("version")["userid"].count()
```

```python
def plot_num_rounds(dfr, ax, show_irrelevant=False):
    for gate in ["gate_30", "gate_40"]:
        ax.plot(pd.DataFrame(dfr[dfr["version"]==gate][
            "sum_gamerounds"].value_counts()).sort_index(), label=gate)

    if show_irrelevant:
        ax.fill_between(np.arange(29), np.arange(29)+3000, color="gray", 
                        alpha=.15, label="irrelevant players")
    ax.legend()
    ax.set_xlabel("Number of Rounds")
    ax.set_ylabel("Number of Players")
```

```python
def plot_cum_num_rounds(dfr, ax, show_irrelevant=False):
    for gate in ["gate_30", "gate_40"]:
        temp_df = dfr[dfr["version"]==gate].copy()
        ax.plot(temp_df.shape[0]-pd.DataFrame(temp_df[
            "sum_gamerounds"].value_counts()).sort_index().cumsum(), label=gate)

    if show_irrelevant:
        ax.fill_between(np.arange(29), np.arange(29)+3000, color="gray", 
                        alpha=.15, label="irrelevant players")
    ax.legend()
    ax.set_xlabel("Number of Rounds")
    ax.set_ylabel("Number of Players")
```

```python
fig, axes = plt.subplots(figsize=(12, 5), ncols=2)

plot_num_rounds(df, axes[0], show_irrelevant=True)
plot_num_rounds(df[df["irrelevant"]==0].copy(), axes[1])

axes[0].set_xlim(0, 100)
axes[0].set_ylim(0, 3000)
axes[0].set_title("All Players")

axes[1].set_xlim(29, 100)
axes[1].set_ylim(0, 350)
axes[1].set_title("Potentially Relevant Players Only")

fig.suptitle("Game Rounds Played during First Two Weeks",  
             y=1.02, fontsize=16)
plt.show()
```

**Key Take-Aways:** 
1. The groups contain roughly the same number of players, even after removing players that are definitely irrelevant.
2. Over 60% of the users in both group did not even reach level 30, and thus there results are not relevant to the experiment. 

**Word of Caution:**

Note that while players who did not play at least 29 rounds are definitely irrelevant to the AB test analysis, many players who did play 29 rounds are also irrelevant.
- <code>sum_gamerounds</code> indicates the *maximum* level the player could have reached; the player did not necessarily reach this level.
    - Players who played 29 or more rounds may have needed attempted to pass a level multiple times. Players who played many rounds are unlikely to be in this category.
- <code>sum_gamerounds</code> is measured after 14 days. <code>retention_1</code> and <code>retention_7</code> are measured after 1 day and 7 days respectively.
    - Players who played 29 or more rounds over the first two weeks may not have played 29 or more rounds by the end of the first week.

Examples of players irrelevant to the analysis who would not be flagged as such:

|sum_gamerounds|sum_gamerounds (by day 7)|level (by day 7)|
|--------------|-------------------------|----------------|
|34|34|26|
|95|18|18|

Number of players assigned to gate_30 version who completed between 29 and 38 rounds:

```python
df[(df["sum_gamerounds"].between(29, 38)) 
& (df["version"]=="gate_30")].shape[0]
```

Number of players assigned to gate_40 version who completed between 29 and 38 rounds:

```python
df[(df["sum_gamerounds"].between(29, 38)) 
   & (df["version"]=="gate_40")].shape[0]
```

It is possible that some players would have lost interest in the game after they played their 29th round and before they completed their 39th round, regardless of which version of the game they were assigned to. Only including these players in the analysis if they were assigned to the gate_40 version could bias the analysis.

# Overall 1-day retention
<p>In the plot above we can see that some players install the game but then never play it (0 game rounds), some players just play a couple of game rounds in their first week, and some get really hooked!</p>
<p>What we want is for players to like the game and to get hooked. A common metric in the video gaming industry for how fun and engaging a game is <em>1-day retention</em>: The percentage of players that comes back and plays the game <em>one day</em> after they have installed it.  The higher 1-day retention is, the easier it is to retain players and build a large player base. </p>
<p>As a first step, let's look at what 1-day retention is overall.</p>

## 1-day retention by AB-group
<p><img src="https://s3.amazonaws.com/assets.datacamp.com/production/project_184/img/belle_cookie.png" style="width:200px; float:right"> </p>
<p>So, a little less than half of the players come back one day after installing the game. Now that we have a benchmark, let's look at how 1-day retention differs between the two AB-groups.</p>

```python
def get_grouped_df(dfr, col):
    gdf = pd.DataFrame(dfr.groupby("version")[
        col].mean()).reset_index()
    #gdf.style.format({col: "{:.1%}"})
    return gdf
```

```python
retention_1_grps = [
    get_grouped_df(df, "retention_1"), 
    get_grouped_df(df[df["irrelevant"]==0], "retention_1"), 
    get_grouped_df(df[df["irrelevant"]==1], "retention_1")]

retention_1_df = retention_1_grps[0].merge(
    retention_1_grps[1], how="inner", on="version").merge(
    retention_1_grps[2], how="inner", on="version")

col_names = {"retention_1_x": "All Players", 
             "retention_1_y": "Potentially Relevant Players", 
             "retention_1": "Irrelevant Players"}
retention_1_df = retention_1_df.rename(
    columns=col_names)

print("Pct of Players Who Played 1st Day After Downloading:")
retention_1_df.style.format({v: "{:.1%}" for v in col_names.values()})
```

## Should we be confident in the difference?
<p>It appears that there was a slight decrease in 1-day retention when the gate was moved to level 40 (44.2%) compared to the control when it was at level 30 (44.8%). It's a small change, but even small changes in retention can have a large impact. But while we are certain of the difference in the data, how certain should we be that a gate at level 40 will be worse in the future?</p>
<p>There are a couple of ways we can get at the certainty of these retention numbers. Here we will use bootstrapping: We will repeatedly re-sample our dataset (with replacement) and calculate 1-day retention for those samples. The variation in 1-day retention will give us an indication of how uncertain the retention numbers are.</p>

```python
import matplotlib.ticker as mtick
```

```python
def plot_bootstrapped_means(dfr, group_var, metric_var, iterations=500):
    # Creating a list with bootstrapped means for each AB-group
    boot = []
    for i in range(iterations):
        boot_mean = dfr.sample(frac=1, replace=True).groupby(
            group_var)[metric_var].mean()
        boot.append(boot_mean)

    boot = pd.DataFrame(boot)
    
    #ax.set_title('Bootstrapped 1-Day Retention Means for each AB-group')
    #ax.set_xlabel('Mean 1-Day Retention')
    # A Kernel Density Estimate plot of the bootstrap distributions
    ax = boot.plot.kde()
    return ax, boot
```

```python
def get_prob_diff(boot_df):
    boot_df["diff"] = ((boot_df["gate_30"] - boot_df["gate_40"]) / 
                       boot_df["gate_40"] * 100)

    # Ploting the bootstrap % difference
    ax = boot_df["diff"].plot.kde()
    ax.set_xlabel("% difference in means")
    #ax.set_title("Bootstrap % Difference in 7-Day Retention Means")

    prob = (boot_df["diff"] > 0).mean()
    
    ax.axvline(color="black", linestyle="--")

    # Pretty printing the probability
    print("Probability that group means are different:", 
          "{0:.1%}".format(prob))
    
    return ax
```

### All Players

```python
rtn1_all_info = plot_bootstrapped_means(df, "version", "retention_1")
my_ax = rtn1_all_info[0]
my_ax.set_title(
    'All Players: \n1st-Day Retention Means by AB-group')
my_ax.set_xlabel('Bootstrapped Mean 1-Day Retention')
my_ax.xaxis.set_major_formatter(mtick.PercentFormatter(1.0, decimals=0))
plt.show()
```

```python
observed_rtn_1_all = pd.crosstab(df["retention_1"], df["version"])
observed_rtn_1_all
```

```python
rtn_1_chi2 = chi2_contingency(observed_rtn_1_all)
```

```python
print("Expected Values:")
pd.DataFrame(rtn_1_chi2[3], columns=["gate_30", "gate_40"]).round(0)
```

```python
print("chi2 p-value:", round(rtn_1_chi2[1], 4))
```

p-value for the chi-squared test less than 0.1, so reject the null hypothesis that 1st-day player retention is equal for both groups. 1st-day player retention was significantly higher in the group of players assigned to the version of the game in which the first gate was placed at level 30.

```python
ttest_ind(df[df["version"]=="gate_30"]["retention_1"].map({True: 1, False: 0}), 
          df[df["version"]=="gate_40"]["retention_1"].map({True: 1, False: 0}))
```

p-value for the t-test less than 0.1, so reject the null hypothesis that 1st-day player retention is equal for both groups. 1st-day player retention was significantly higher in the group of players assigned to the version of the game in which the first gate was placed at level 30.

```python
fig, my_ax = plt.subplots()
my_ax = get_prob_diff(rtn1_all_info[1])
my_ax.set_title("Differences in Group Means")
plt.show()
```

### Potentially Relevant Players

```python
rtn1_potrel_info = plot_bootstrapped_means(
    df[df["irrelevant"]==0], "version", "retention_1")
my_ax = rtn1_potrel_info[0]
my_ax.set_title(
    'Potentially Relevant Players: \n1st-Day Retention Means by AB-group')
my_ax.set_xlabel('Bootstrapped Mean 1-Day Retention')
my_ax.xaxis.set_major_formatter(mtick.PercentFormatter(1.0, decimals=0))
plt.show()
```

```python
observed_rtn_1_potrel = pd.crosstab(
    df[df["irrelevant"]==0]["retention_1"], 
    df[df["irrelevant"]==0]["version"])
observed_rtn_1_potrel
```

```python
rtn_1_potrel_chi2 = chi2_contingency(observed_rtn_1_potrel)
```

```python
print("Expected Values:")
pd.DataFrame(rtn_1_potrel_chi2[3], columns=["gate_30", "gate_40"]).round(0)
```

```python
print("chi2 p-value:", round(rtn_1_potrel_chi2[1], 4))
```

p-value for the chi-squared test is greater than than 0.1, so we fail to reject the null hypothesis that 1st-day player retention is equal for both groups.

```python
ttest_ind(df[(df["irrelevant"]==0) & (df["version"]=="gate_30")
            ]["retention_1"].map({True: 1, False: 0}), 
          df[(df["irrelevant"]==0) & (df["version"]=="gate_40")
            ]["retention_1"].map({True: 1, False: 0}))
```

p-value for the t-test is greater than than 0.1, so we fail to reject the null hypothesis that 1st-day player retention is equal for both groups.

```python
fig, my_ax = plt.subplots()
my_ax = get_prob_diff(rtn1_potrel_info[1])
my_ax.set_title("Differences in Group Means")
plt.show()
```

# Overall 7th-Day Retention

## 7th-day retention by AB-group
<p>The bootstrap analysis tells us that there is a high probability that 1-day retention is better when the gate is at level 30. However, since players have only been playing the game for one day, it is likely that most players haven't reached level 30 yet. That is, many players won't have been affected by the gate, even if it's as early as level 30. </p>
<p>But after having played for a week, more players should have reached level 40, and therefore it makes sense to also look at 7-day retention. That is: What percentage of the people that installed the game also showed up a week later to play the game again.</p>
<p>Let's start by calculating 7-day retention for the two AB-groups.</p>

```python
retention_7_grps = [
    get_grouped_df(df, "retention_7"), 
    get_grouped_df(df[df["irrelevant"]==0], "retention_7"), 
    get_grouped_df(df[df["irrelevant"]==1], "retention_7")]

retention_7_df = retention_7_grps[0].merge(
    retention_7_grps[1], how="inner", on="version").merge(
    retention_7_grps[2], how="inner", on="version")

col_names = {"retention_7_x": "All Players", 
             "retention_7_y": "Potentially Relevant Players", 
             "retention_7": "Irrelevant Players"}
retention_7_df = retention_7_df.rename(
    columns=col_names)

print("Pct of Players Who Played 7th Day After Downloading:")
retention_7_df.style.format({v: "{:.1%}" for v in col_names.values()})
```

### All Players

<p>Like with 1-day retention, we see that 7-day retention is slightly lower (18.2%) when the gate is at level 40 than when the gate is at level 30 (19.0%). This difference is also larger than for 1-day retention, presumably because more players have had time to hit the first gate. We also see that the <em>overall</em> 7-day retention is lower than the <em>overall</em> 1-day retention; fewer people play a game a week after installing than a day after installing.</p>
<p>But as before, let's use bootstrap analysis to figure out how certain we should be of the difference between the AB-groups.</p>

```python
rtn7_all_info = plot_bootstrapped_means(df, "version", "retention_7")
my_ax = rtn7_all_info[0]
my_ax.set_title(
    'All Players: \n7th-Day Retention Means by AB-group')
my_ax.set_xlabel('Bootstrapped Mean 7th-Day Retention')
my_ax.xaxis.set_major_formatter(mtick.PercentFormatter(1.0, decimals=0))
plt.show()
```

```python
observed_rtn_7_all = pd.crosstab(df["retention_7"], df["version"])
observed_rtn_7_all
```

```python
rtn_7_chi2 = chi2_contingency(observed_rtn_7_all)
```

```python
print("Expected Values:")
pd.DataFrame(rtn_7_chi2[3], columns=["gate_30", "gate_40"]).round(0)
```

```python
print("chi2 p-value:", round(rtn_7_chi2[1], 4))
```

p-value for the chi-squared test less than 0.1, so reject the null hypothesis that 1st-day player retention is equal for both groups. 1st-day player retention was significantly higher in the group of players assigned to the version of the game in which the first gate was placed at level 30.

```python
ttest_ind(df[df["version"]=="gate_30"]["retention_7"].map({True: 1, False: 0}), 
          df[df["version"]=="gate_40"]["retention_7"].map({True: 1, False: 0}))
```

p-value for the t-test less than 0.1, so reject the null hypothesis that 1st-day player retention is equal for both groups. 1st-day player retention was significantly higher in the group of players assigned to the version of the game in which the first gate was placed at level 30.

```python
fig, my_ax = plt.subplots()
my_ax = get_prob_diff(rtn7_all_info[1])
my_ax.set_title("Differences in Group Means")
plt.show()
```

# 7th Day Retention Given 1st Day Retention

## 1st-day retention follow-up by AB-group

<p>The bootstrap analysis tells us that there is a high probability that 1-day retention is better when the gate is at level 30. However, since players have only been playing the game for one day, it is likely that most players haven't reached level 30 yet. That is, many players won't have been affected by the gate, even if it's as early as level 30. </p>
<p>But after having played for a week, more players should have reached level 40, and therefore it makes sense to also look at 7-day retention. That is: What percentage of the people that installed the game also showed up a week later to play the game again.</p>
<p>Let's start by calculating 7-day retention for the two AB-groups.</p>

```python
df_d1r = df[df["retention_1"]].copy()
```

```python
retention_7fou_grps = [
    get_grouped_df(df_d1r, "retention_7"), 
    get_grouped_df(df_d1r[df_d1r["irrelevant"]==0], "retention_7"), 
    get_grouped_df(df_d1r[df_d1r["irrelevant"]==1], "retention_7")]

retention_7fou_df = retention_7fou_grps[0].merge(
    retention_7fou_grps[1], how="inner", on="version").merge(
    retention_7fou_grps[2], how="inner", on="version")

col_names = {"retention_7_x": "All Players", 
             "retention_7_y": "Potentially Relevant Players", 
             "retention_7": "Irrelevant Players"}
retention_7fou_df = retention_7fou_df.rename(
    columns=col_names)

print("Pct of the Players Who Played 1st Day After Downloading \
Who Also Played 7th Day:")
retention_7fou_df.style.format({v: "{:.1%}" for v in col_names.values()})
```

```python
ct_g30 = pd.crosstab(
    df[df["version"]=="gate_30"]["retention_1"],
    df[df["version"]=="gate_30"]["retention_7"])
ct_g30 = ct_g30/df[df["version"]=="gate_30"].shape[0]
ct_g30.style.format({False: "{:.1%}", True: "{:.1%}"})
```

```python
ct_g40 = pd.crosstab(
    df[df["version"]=="gate_40"]["retention_1"],
    df[df["version"]=="gate_40"]["retention_7"])
ct_g40 = ct_g40/df[df["version"]=="gate_40"].shape[0]
ct_g40.style.format({False: "{:.1%}", True: "{:.1%}"})
```
